## Usinn KNN alogrithm to perform ML on the datset to know whether a person will have diabetes or not

```python

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix # testing the models downwards
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
```

```python
df.head()
```

```python
# split the dataset
X = df.iloc[:,0:8]
y = df.iloc[:,8]
X_train,X_test,y_train,y_test = train_test_split(X,y,  random_state=0, test_size=0.2)
```

```python
# feature scaling
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
import math
math.sqrt(len(y_test))
```

```python
# Define te model :Init kNN
classifier = KNeighborsClassifier(n_neighbors=11,p=2,metric='euclidean')
```

```python
print(X.isnull().sum())  # Check how many NaN values exist per column
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy="mean")  # Options: "mean", "median", "most_frequent"
X_train = imputer.fit_transform(X_train)
X_test = imputer.transform(X_test)  # Apply the same transformation to X_test
```

```python
classifier.fit(X_train, y_train) 
```

```python
# using the confusion matrix for evaluating the models
# Predict the x_test results
y_pred = classifier.predict(X_test)
y_pred
```

```python
# Evaluate the model
cm = confusion_matrix(y_test,y_pred)
print (cm)
# diagonol 92 and 27  were classified correctly while the diagonal 20  15 had diabetes and 20 were wrongly classifed meaning
```

```python
print(f1_score(y_test,y_pred))
```

```python
print(accuracy_score(y_test,y_pred))
```
